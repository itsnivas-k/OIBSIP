```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#pandas is a library function used for working with datasets
#numpy is a library function used for woring with arrays
#Matplotlib is a library function for creating static, animated, and interactive visualizations in Python
#Seaborn library is a used for data visualization 
# Python library is an  plotting library used for making chart types for  statistical, financial, geographic, scientific, and 3-dimensional use-cases.


data = pd.read_csv("https://raw.githubusercontent.com/itsnivas-k/OIBSIP/main/Unemployment_Rate_upto_11_2020.csv")
print(data.head())
#dataset has been feeded  
```

```python
#This displays the contents of the dataset upto first five values
```

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
```

```python
#This is used to fetch the columns from the dataset
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(data.corr())
plt.show()
#Using Seaborn Library function heatmap will be generated along with its coordinates
#plt.show() is used to display it 
```

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.show()
#This is used to fetch the columns from the dataset
#Histplot is used to display the data in a form of histogram
#This block displays the Estimated Employed Count along withn thier Region
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
#This is used to fetch the columns from the dataset
#Histplot is used to display the data in a form of histogram
#This block displays the Estimated Unemployment Rate Count along withn thier Region in a histogram
```

```python
unemploment = data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region", "States"], 
                     values="Estimated Unemployment Rate", 
                     width=700, height=700, color_continuous_scale="RdBu", 
                     title="Unemployment Rate in India")
figure.show()
#sunburst chart is used  for displaying hierarchical data
#This block displays the Estimated Unemployment Rate Count along withn thier Region in a sunburst chart type
```
